```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
#location of csv files
crime_file = "C:\\Users\\BGaither\\OneDrive - McAfee\\Other\\school\\SMU\\Courses\\Fall 2020\\Machine Learning I\\Lab1\\Data_NYPD\\Modified_Data_Sets\\FE1_NYPD_Complaint_Data_Historic.csv"
weather_file = "C:\\Users\\BGaither\\Documents\\GitHub\\NewYorkCityCrimes2015\\Data\\NYC_Weather_2010-2020.csv"
income_file = "C:\\Users\\BGaither\\Documents\\GitHub\\NewYorkCityCrimes2015\\Data\\NYC_per_capita_income_2014_2015.csv"
housing_file = "C:\\Users\\BGaither\\Documents\\GitHub\\NewYorkCityCrimes2015\\Data\\Housing_Prices_Cleaned.csv"
population_file = "C:\\Users\\BGaither\\Documents\\GitHub\\NewYorkCityCrimes2015\\Data\\Population_by_Borough_NYC.csv"
#load Data into Pandas
df_Crime = pd.read_csv(crime_file)
df_Weather = pd.read_csv(weather_file)
df_Income = pd.read_csv(income_file)
df_Housing = pd.read_csv(housing_file)
df_Pop = pd.read_csv(population_file)
```

```python
#check the shape of the original crime data
df_Crime.shape
```

```python
#this will select only the top 6 crimes by volume

#ASSAULT 3 & RELATED OFFENSES -- 344
#CRIMINAL MISCHIEF & RELATED OF -- 351,121
#DANGEROUS DRUGS -- 117,235
#GRAND LARCENY -- 109
#HARRASSMENT 2 -- 578
#PETIT LARCENY -- 341

df_Crime = df_Crime[df_Crime.KY_CD.isin([344,351,121,117,235,109,578,341])]
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df_aggregated = df_Crime.groupby(['CMPLNT_FR_DT','Daytime','Day_Name','Month', 'Day','Year','Season','GeoCell','BORO_NM']).agg(count_cmplnt=('CMPLNT_NUM','count')).reset_index()
df_aggregated.head(5)
```

```python
df_crime_weather = df_aggregated.merge(df_Weather,left_on='CMPLNT_FR_DT',right_on='DATE',how='inner')[['CMPLNT_FR_DT','Daytime','Day_Name','Month', 'Day','Year','Season','GeoCell','BORO_NM','count_cmplnt','PRCP','SNOW','TMIN','TMAX']]
df_crime_weather.head(5)
```

```python
#df_Pop = pd.read_csv(population_file)
df_Pop = df_Pop[['Borough','2010']]
df_Pop = df_Pop[df_Pop.Borough != 'NYC Total']
df_Pop['Borough'] = df_Pop['Borough'].str.upper()
df_Pop = df_Pop.rename(columns ={'2010':'Population'})
df_Pop['Borough'] = df_Pop['Borough'].str.strip()
df_Pop
```

```python
df_cime_pop_weather = df_crime_weather.merge(df_Pop,left_on='BORO_NM',right_on='Borough',how='inner')[['CMPLNT_FR_DT','Daytime','Day_Name','Month', 'Day','Year','Season','GeoCell','BORO_NM','count_cmplnt','PRCP','SNOW','TMIN','TMAX','Population']]
df_cime_pop_weather.head(5)
```

```python
#df_Income
df_crime_pop_weather_inc = df_cime_pop_weather.merge(df_Income,left_on=['BORO_NM','Year'],right_on=['BORO_NM','YEAR'],how='inner')[['CMPLNT_FR_DT','Daytime','Day_Name','Month', 'Day','Year','Season','GeoCell','BORO_NM','count_cmplnt','PRCP','SNOW','TMIN','TMAX','Population','PC_INCOME']]
df_crime_pop_weather_inc.head(5)
```

```python
df_crime_pop_weather_inc_house = df_crime_pop_weather_inc.merge(df_Housing,left_on=['BORO_NM','Year'],right_on=['BOROUGH','YEAR'])[['CMPLNT_FR_DT','Daytime','Day_Name','Month', 'Day','Year','Season','GeoCell','BORO_NM','count_cmplnt','PRCP','SNOW','TMIN','TMAX','Population','PC_INCOME','Hm_Sls_Price_Range']]
df_crime_pop_weather_inc_house.head(5)
```

```python
#output modified dataframe to csv for later use
df_crime_pop_weather_inc_house.to_csv(r'C:\\Users\\BGaither\\OneDrive - McAfee\\Other\\school\\SMU\\Courses\\Fall 2020\\Machine Learning I\\Lab1\\Data_NYPD\\Modified_Data_Sets\\NYPD_Rollup_Joined_All.csv',index=False,header=True)
```
